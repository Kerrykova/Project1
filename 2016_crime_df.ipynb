```python
import requests
import json
import pandas as pd

url = "https://data.austintexas.gov/resource/qxhx-pr9e.json"
```

```python
response = requests.get(url).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
crime_df_2016 = pd.DataFrame(response)
crime_df_2016
```

```python
crime_df_2016.to_csv("2016_crime_data.csv", encoding="utf-8", index=False)
```
