```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from zipcode import zipcodes
```

```python
zipcode = [78610, 78613, 78617, 78641, 78652, 78653, 78660, 78664, 78681, 78701, 78702, 78703, 78704, 78705, 78712, 78717, 78719, 78721, 78722, 78723, 78724, 78725, 78726, 78727, 78728, 78729, 78730, 78731, 78732, 78733, 78734, 78735, 78736, 78737, 78738, 78739, 78741, 78742, 78744, 78745, 78746, 78747, 78748, 78749, 78750, 78751, 78752, 78753, 78754, 78756, 78757, 78758, 78759]

census_2011 = pd.read_csv("Raw Data/raw_census_data_2011.csv")

df = pd.DataFrame
df = census_2011.set_index("Zipcode")


new_census_2011 = []


for x in zipcode:
    a=df.loc[x]
    new_census_2011.append(a)
      
new_census_2011 = pd.DataFrame(new_census_2011)

new_census_2011 = new_census_2011[["Population", "Median Age", "Household Income", "Per Capita Income"]]
new_census_2011.index.name = "Zipcodes"
new_census_2011.head()

```

```python
        
census_2016 = pd.read_csv("Raw Data/raw_census_data_2016.csv")

df2 = pd.DataFrame
df2 = census_2016.set_index("Zipcode")


new_census_2016 = []


for x in zipcode:
    b=df2.loc[x]
    new_census_2016.append(b)
      
new_census_2016 = pd.DataFrame(new_census_2016)
new_census_2016 = new_census_2016[["Population", "Median Age", "Household Income", "Per Capita Income"]]
new_census_2016.index.name = "Zipcodes"
new_census_2016.head()


   
```

```python
# census= new_census_2011.reset_index().merge(new_census_2016, how="left").set_index('index')
census = pd.merge(new_census_2011, new_census_2016, on="Zipcodes", how = "outer", suffixes=("_2011","_2016"))
census.head()

# census = pd.DataFrame({'Income Increase/Decrease': (new)})
```

```python
census["PC Income Increase/Decrease"] = ""
census["PC % Income Increase/Decrease"] = ""
census["HH Income Increase/Decrease"] = ""
census["HH % Income Increase/Decrease"] = ""
census.head()
```

```python
census["PC Income Increase/Decrease"] = census["Per Capita Income_2016"] - census["Per Capita Income_2011"]
census["PC % Income Increase/Decrease"] = census["PC Income Increase/Decrease"] / census["Per Capita Income_2011"] * 100
census["HH Income Increase/Decrease"] = census["Household Income_2016"] - census["Household Income_2011"]
census["HH % Income Increase/Decrease"] = census["HH Income Increase/Decrease"] / census["Household Income_2011"] * 100
census.head()
```
