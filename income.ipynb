```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from census import Census
from us import states

# Census API Key
from config import api_key
api_key
```

```python
#2016 api pull
c_2016 = Census(api_key, year=2016)
census_data_2016 = c_2016.acs5.get(("NAME", "B19013_001E", "B01003_001E", "B01002_001E",
                          "B19301_001E"), 
                         {'for': 'zip code tabulation area:*'})
```

```python
#2016 api to dataframe
census_df_2016 = pd.DataFrame(census_data_2016)
census_df_2016 = census_df_2016.rename(columns={
    "zip code tabulation area": "Zipcode",
                                      "NAME": "Name", 
                                      "B01003_001E": "Population",
                                      "B01002_001E": "Median Age",
                                      "B19013_001E": "Household Income",
                                      "B19301_001E": "Per Capita Income"
                                     })
census_df_2016 = census_df_2016[["Zipcode", "Population", "Median Age", "Household Income", "Per Capita Income"]]
census_df_2016.head()
```

```python
#raw 2016 dataframe saved as .csv
census_df_2016.to_csv("raw_census_data_2016.csv")
```
