```python
import pandas as pd

crime_df = pd.read_csv("/Users/delaynabradshaw/Desktop/Crime_Reports__beta_.csv")
crime_df.head()
```

```python
#Narrow the dataframe to only the information we are interested in 
narrowed_crime = crime_df.drop(['Report Date', 'Report Time', 'Location Type', 'Address', 'Council District', 'APD Sector', 'APD District', 'PRA','Incident Number', 'Highest Offense Code', 'Family Violence', 'Occurred Date Time', 'Occurred Time', 'Report Date Time', 'Census Tract', 'Clearance Status', 'Clearance Date', 'UCR Category', 'Category Description', 'X-coordinate', 'Y-coordinate', 'Latitude', 'Longitude', 'Location'], axis=1)
narrowed_crime.head()
```

```python
#Added a year column
narrowed_crime['Occurred Date'] = pd.to_datetime(narrowed_crime['Occurred Date'])
narrowed_crime['Year'] = narrowed_crime['Occurred Date'].dt.year
narrowed_crime.head()
```

```python
#Drop the Occurred Date column
narrowed_crime = narrowed_crime.drop(['Occurred Date'], axis=1)
narrowed_crime.head()
```

```python
#we want to search years and find only 2011 and 2016
years_2011 = narrowed_crime.loc[narrowed_crime["Year"] == 2011]
years_2011.head()
```

```python
years_2016 = narrowed_crime.loc[narrowed_crime["Year"] == 2016]
years_2016.head()
```

```python
#groupby zipcode 2011 
groupedby_zipcode = years_2011.groupby('Zip Code')

#count number of offenses per zipcode
count_2011_zip_offenses = groupedby_zipcode['Zip Code'].count()
count_2011_zip_offenses
```

```python
len(count_2011_zip_offenses)
```

```python
#groupedby_zipcode_type = years_2011.groupby('Zip Code', 'Highest Offense Description')
count_zipcode_type_2011_unfinished = years_2011.groupby(['Zip Code', 'Highest Offense Description'], as_index=False).count()

count_zipcode_type_2011 = count_zipcode_type_2011_unfinished.rename(columns={'Year':'Number of Occurances'})

count_zipcode_type_2011.to_pickle("Intermediate Data/count_zipcode_type_2011.pkl")

count_zipcode_type_2011.head()
```

```python
#groupby zipcode 2016 and count number of offenses
groupedby_zipcode = years_2016.groupby('Zip Code')

count_2016_zip_offenses = groupedby_zipcode['Zip Code'].count()
count_2016_zip_offenses
```

```python
len(count_2016_zip_offenses)
```

```python
#groupedby_zipcode_type = years_2011.groupby('Zip Code', 'Highest Offense Description')
count_zipcode_type_2016_unfinished = years_2016.groupby(['Zip Code', 'Highest Offense Description'], as_index=False).count()

count_zipcode_type_2016 = count_zipcode_type_2016_unfinished.rename(columns={'Year':'Number of Occurances'})

count_zipcode_type_2016.to_pickle("Intermediate Data/count_zipcode_type_2016.pkl")

count_zipcode_type_2016.head()
```
